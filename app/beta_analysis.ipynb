```python
def filename(x):
    return f"beta_analysis/result_2_{x}.flight_stats.csv"
```

```python
import pandas as pd

data = {}

for j in range(10,90):
    try:
        d = pd.read_csv(filename(j))
        i = j / 100
        data[i] = {}
        data[i]["upgraded_pnr"] = d["upgraded_pnr"].sum()
        data[i]["samestate_pnr"] = d["samestate_pnr"].sum()
        data[i]["downgraded_pnr"] = d["downgraded_pnr"].sum()
        data[i]["unallocated_pnr"] = d["unallocated_pnr"].sum()
        data[i]["allocated_pnr"] = d["allocated_pnr"].sum()
        data[i]["allocated_passengers"] = d["allocated_passengers"].sum()
        data[i]["unallocated_passengers"] = d["unallocated_passengers"].sum()
    except:
        pass

# for key, value in data.items():
#     print(key, value)
```

```python
import matplotlib.pyplot as plt

# Assuming you have already processed the data and stored it in the 'data' dictionary

# Create a DataFrame from the 'data' dictionary
df = pd.DataFrame.from_dict(data, orient='index')

# Plot parallel line chart
plt.figure(figsize=(10, 6))
for column in df.columns:
    plt.plot(df.index, df[column], label=column)

plt.xlabel('Key (or i)')
plt.ylabel('Values')
plt.title('Parallel Line Chart')
plt.legend()
plt.show()
```
